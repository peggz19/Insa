```python
#Dependencies
import pandas as pd
import json
import requests
from pathlib import Path
```

```python
#Reference files
followers_file_path=Path('followers_1.json')
followers_content=followers_file_path.read_text()
followers_data=json.loads(followers_content)
followers_data[32]
```

```python
#Creating a list in which to make each insta id
followers_list =[]
#followers_data[0]["string_list_data"][0]["value"] helped finding where the id is
#len(followers_data)
for i in range(len(followers_data)):
    #print(followers_data[i]['string_list_data'][i]["value"])
    followers_list.append(followers_data[i]['string_list_data'][0]["value"])
followers_list[1:5]
```

```python
#Reference files
following_file_path=Path('following.json')
following_content=following_file_path.read_text()
following_data=json.loads(following_content)
following_data
```

```python
#Creating a list
following_list = []
following_data['relationships_following'][0]["string_list_data"][0]['value']
len(following_data['relationships_following'])
for i in range(len(following_data["relationships_following"])):
    following_list.append(following_data["relationships_following"][i]['string_list_data'][0]["value"])
following_list[1:5]
```

```python
#Creating a dataframe with the list above
df=pd.DataFrame({"Following":following_list})
df.head(5)
```

```python
#Now create a rule that says Yes or No if following me
follows_me = []
for i in following_list:
    if i in followers_list:
        follows_me.append("yes")
    else:
        follows_me.append("no")
follows_me[0:5]
```

```python
#Now we merge the above to pandas
df["Follows Me"]=follows_me
df.head(5)
```

```python
#Export to Excel
#len(following_data["relationships_following"])
df.to_excel("output.xlsx")
```
